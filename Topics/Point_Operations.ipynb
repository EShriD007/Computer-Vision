```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Point Operations
These are operations that directly alter each pixels independent of the neighbours.\
s = T(r) - s. Here s is the output pixel, T is the transformation and r is the input pixel.\
Thresholding, brightness increasing are examples of Point Operations

## Other Examples
1. **Image Negation**
s = (Max Intensity) - r This will make all the dark pixels into light and vice versa.
2. **Log Transform**
s = c * log(1 + r) This will increase the dark region's visibility and will reduce the bright region's glare. (This will reduce the spread of the bright pixels and will increase the intensity of the dark pixels)
3. **Power Law Transform**
s = c * r^Gamma\
If Gamma < 1 -> Expand Dark Region in the Histogram (Increase the Brightness of the image)\
If Gamma > 1 -> Compress Dark region in the Histogram (Decrease the Brightness of the Image)\
If Gamma = 1 -> No changes.

```python
# Compute Negative
bgr_img = cv2.imread("images/otsu_equations.png", cv2.IMREAD_COLOR)
rgb_color = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
rgb_color_neg = 255 - rgb_color

plt.imshow(rgb_color)
plt.title("Color Image")
plt.show()
plt.imshow(rgb_color_neg)
plt.title("Color Negative")
plt.show()
```

```python
# Compute Log Transform - s = c * log(1 + r)
# Convert the image to float
rgb_color_float = rgb_color.astype(np.float32)
c = 255 / (np.log(1 + np.max(rgb_color_float)))
rgb_color_log = c * np.log(1 + rgb_color_float)

#Power Law Transform
gamma = 2
rgb_color_power = c * rgb_color_float ** gamma

#Normalize within 0 to 255 for display
rgb_color_log_norm = cv2.normalize(rgb_color_log, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
rgb_color_power_norm = cv2.normalize(rgb_color_power, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

plt.imshow(rgb_color_log_norm)
plt.title("Log Image")
plt.show()

plt.imshow(rgb_color_power_norm)
plt.title("Power Law Image")
plt.show()
```
